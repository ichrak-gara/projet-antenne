```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('test_df.csv',  delimiter=';', encoding='latin-1')
df.tail()
```

```python
df.rename(columns={'Niveau global': 'Niveau_global'}, inplace=True)
```

```python
df=df.assign(target=1)
print(df)
```

```python
df.columns
```

```python
def target():
    for t in range(0,len(df)):
        if (df['Niveau_global'][t]>6):
            df['target'][t] =0
            print(df['Niveau_global'][t])
```

```python
target()
```

```python
pd.crosstab(df.target, "freq")
```

```python
df.tail(30)
```

```python
X=df[['Numéro de mesure','Niveau_global']].to_numpy()
print(X)
y=df['target'].to_numpy()
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
#create a l.reg classifier
logreg1 = LogisticRegression()
#train the model using the training sets
logreg1.fit(X_train, y_train)
y_pred = logreg1.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
